some players that played for multiple teams throughout one regular season have multiple entries in the same year for each team they played for , we need to remove this and ony show each player once each year, example : greg anderson

```python
 players.groupby(['Player','Year']).get_group(('Greg Anderson',1991))
```

we will write a function that removes multiple instances and only return the TOT row, this is their total years stats, so we pass the group of each player to this function then if the number of rows in this group is more than 1 we return the row that has team name as TOT

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row=df[df['Tm']=='TOT']
        row['Tm']=df.iloc[-1,:]['Tm']#replaces TOT with the last team the player played for 
        return row

players=players.groupby(['Player','Year']).apply(single_row)

#this makes groups of each player and year, players with multiple
#entries in the same year get grouped together 
```

```python
players.index=players.index.droplevel()
```

```python
players.index=players.index.droplevel()
```

both data sets have different notations for team names , teams df has full team names while players combined data set has shorteneed versions of team names


```python
nicknames={}

with open('nicknames.csv')as f:
    lines=f.readlines()
    for line in lines[1:]:
        abbrev,name=line.replace('\n','').split(',')
        nicknames[abbrev]=name
```

```python
combined['Team']=combined['Tm'].map(nicknames)
```
