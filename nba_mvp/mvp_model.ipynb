we will use the stats of players to predct how many mvp points they would win , or more precisly what share of mvp votes they get , this would help determining wether the player wins or not at the end

```python
import pandas as pd
```

```python
stats =pd.read_csv('player_mvp_stats.csv')
```

```python
stats
```

```python
del stats['Unnamed: 0']
```

```python
pd.isnull(stats).sum()
```

the null values are present in 3pt percentage and ft percentage , this could be because a player has not attempted any 3pt shots in the season so there is no value for that number , we could replace them wiht 0 instead if this is the case 

```python
stats[pd.isnull(stats['3P%'])][['Player','3PA']]
```

we can see that when 3pa is 0 , 3p% is null, lets check it for ft percent

```python
stats[pd.isnull(stats['FT%'])][['Player','FTA']]
```

now replace all null values with zero

```python
stats=stats.fillna(0)
```

```python
stats.info()
```

```python
stats.isnull().sum()
```

now all null values have been removed


```python
stats.columns
```

we will use numeric attributes to predict mvp, remove attributes like team , player name position etc , and also remove the stats about mvp shares won from the predictors set as they are the target variables

```python
predictors=['Age','G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year','W', 'L', 'W/L%', 'GB', 'PS/G',
       'PA/G', 'SRS']
```

```python
train = stats[stats["Year"]<2022]
test = stats[stats['Year']==2022]
```

ridge regression which is a modification of linear regression will be used here
this reduces overfitting due to multi collinearity of attributes by introducing a regularization parameter while calculating sum of squared error

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.1)    
```

```python
reg.fit(train[predictors],train['Share'])
```

```python
predictions=reg.predict(test[predictors])
```

```python
predictions=pd.DataFrame(predictions,columns=['predictions'],index=test.index)
```

```python
predictions
```

```python
combination=pd.concat([test[['Player','Share']],predictions],axis=1)
```

```python
combination
```

```python
combination.sort_values("Share",ascending=False).head(10)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(combination['Share'],combination['predictions'])
```

this isnt a good error metric as we dont actualy care about the actual number instead we need the ranking of players acc to this metric , also a vast majority of values in this are 0 so it would make using mean squared error not suitable as it calculates the raw differnce in values squared

we dont care if the model predicts the bottom 380 or so players mvp shares as they would be zero ,most of the players do not edn  up receiving any votes , we want our model to be accurate for the top players

```python
combination=combination.sort_values('Share',ascending=False)
combination['Rk']=list(range(1,combination.shape[0]+1))
# adding rank column to combinations df and also sorting them in order of share
```

```python
combination.head(10)
```

```python
combination=combination.sort_values('predictions',ascending=False)
combination['Predicted Rk']=list(range(1,combination.shape[0]+1))
# sorting by predicted share value
```

```python
combination.head(10)
```

i want to create an error metric that sees if the actual top 5 mvp finishers were predicted in the top 5 , 

for players correctly ranked there is no penalty, but if not, we see how further down they are predicted than their actual ranking , and the larger this is , the more will be the penalty
