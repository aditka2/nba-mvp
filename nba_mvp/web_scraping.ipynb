first i will scrape data from basketball reference , to get data on previous years mvp winners and mvp points for the other players in the running for the award.


```python
years=list(range(1991,2023))
print(years,end="")
```

```python
url_start='https://www.basketball-reference.com/awards/awards_{}.html'
```

```python
import requests

for year in years:
    url=url_start.format(year) #replaces {} with specific year
    data=requests.get(url)
    
    with open('mvps/{}.html'.format(year),'w+',encoding='utf-8') as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open('mvps/1991.html',encoding='utf-8') as f:
    page=f.read()
```

```python
soup=BeautifulSoup(page,"html.parser")
```

the mvp stats table has a header over the individual stats saying voting per game shooting and advanced , we dont want this at it will complicate the df where we extract data to , so we will find this by using soup.find and remove it from the table 

then we will find the actual class of the table in the html file and extract data from that 

the over header has a tr tag  with the class name over_header so we will find and remove this 

```python
soup.find('tr',class_='over_header').decompose()
```

by looking at the html page , the table we want an element with table tag, that has a id='mvp' and class ='sortable stats_table

```python
mvp_table=soup.find(id='mvp')
mvp_table
```

now we have the table data as an html table stored in mvp_table variable, pandas has the ability to read from html tables , so this is  how we will extract the tables into a df

```python
import pandas as pd
```

```python
mvp_1991=pd.read_html(str(mvp_table))[0]
```

```python
mvp_1991
```

pd.read_html is giving us a list of data frames with one table in the list , we will take the first element of the list to get the dataframe thaty we want

next we will make a loop that does this for the data of every season in our list and make a list of dataframes for each years mvp table

```python
dfs=[]
for year in years:
    with open('mvps/{}.html'.format(year),encoding='utf-8') as f:
     page=f.read()
    soup=BeautifulSoup(page,"html.parser")
    soup.find('tr',class_='over_header').decompose()
    mvp_table=soup.find(id='mvp')
    mvp=pd.read_html(str(mvp_table))[0]
    mvp['Year']=year
    dfs.append(mvp)
```

```python
mvps=pd.concat(dfs)
```

```python
mvps.head()
```

all the years mvp dfs were combined into one large df, and i also added a column that shows the year , since that data is lost whe we combine 

```python
mvps.to_csv('mvps.csv')
```

on the webpage there are some rows that reiterate the categories of the columns every 40 players or so, player pos age ....., we only need this to appear once in the dataset ,so we will find this table row element and remove it using beatuifulsoup ,leaving only the stats table that we want 

from inspecting the page i can see that , these rows are of tr tag and class='thead', so now i will remove all instances of this except the first one at the very top

```python
year=1991
with open("players/{}.html".format(year),encoding='utf-8')as f:
    page=f.read()
soup=BeautifulSoup(page,"html.parser")
soup.find('tr',class_='thead').decompose()
player_table=soup.find(id='per_game_stats')
player=pd.read_html(str(player_table))[0]
player['Year']=year
```

```python
years=list(range(2010,2022))

for year in years:
    
    url='https://www.basketball-reference.com/leagues/NBA_{}_standings.html'.format(year)

    data=requests.get(url)

    with open('team/{}.html'.format(year),'w+',encoding='utf-8') as f:
        f.write(data.text)
```

```python
dfs=[]
years=list(range(1991,2023))
for year in years:
    with open('team/{}.html'.format(year),encoding='utf-8')as f:
        page=f.read()
    soup=BeautifulSoup(page,'html.parser')
    soup.find('tr',class_='thead').decompose()
    team_table=soup.find(id='divs_standings_E')
    team=pd.read_html(str(team_table))[0]
    team['Year']=year
    team["Team"]=team["Eastern Conference"]
    del team["Eastern Conference"]
    dfs.append(team)

    soup=BeautifulSoup(page,'html.parser')
    soup.find('tr',class_='thead').decompose()
    team_table=soup.find(id='divs_standings_W')
    team=pd.read_html(str(team_table))[0]
    team['Year']=year
    team["Team"]=team["Western Conference"] 
    del team["Western Conference"]
    dfs.append(team)
    
```

```python
teams=pd.concat(dfs)
```
